```python
import os
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
%matplotlib inline
from keras.applications.xception import Xception, preprocess_input
from keras.preprocessing.image import load_img, img_to_array
```

```python
import glob
```

```python
os.getcwd()
```

```python
data_dir = r"C:\Users\ABC\Documents\IBM AI 6th sem\2. Deep Learning\11. CNN 13-03\15 Convolutional Neural Network\tranfer learning with logistic regression"
```

```python
def preprocess(img):
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array,axis=0)
    img_preprocess = preprocess_input(img_array)
    return img_preprocess[0]
```

```python
os.chdir(os.path.join(data_dir,"train","waffles"))
waffles_train_image_names = glob.glob("*.png")
```

```python
waffles_train=[]
for i in waffles_train_image_names:
    img = load_img(i,target_size=(224,224))
    waffles_train.append(preprocess(img))
```

```python
os.chdir(os.path.join(data_dir,"train","pancakes"))
pancakes_train_image_names = glob.glob("*.png")
```

```python
pancakes_train = []
for i in pancakes_train_image_names:
    img = load_img(i,target_size=(224,224))
    pancakes_train.append(preprocess(img))
```

```python
os.chdir(os.path.join(data_dir,"test","waffles"))
waffles_test_image_names = glob.glob("*.png")
```

```python
waffles_test=[]
for i in waffles_test_image_names:
    img = load_img(i,target_size=(224,224))
    waffles_test.append(preprocess(img))
```

```python
os.chdir(os.path.join(data_dir,"test","pancakes"))
pancakes_test_image_names = glob.glob("*.png")
```

```python
pancakes_test = []
for i in pancakes_test_image_names:
    img = load_img(i,target_size=(224,224))
    pancakes_test.append(preprocess(img))
```

```python
print(len(waffles_train))
print(len(pancakes_train))
print(len(waffles_test))
print(len(pancakes_test))
```

```python
waffles_train = np.array(waffles_train,dtype='float32')
waffles_test = np.array(waffles_test,dtype='float32')
```

```python
pancakes_train = np.array(pancakes_train,dtype='float32')
pancakes_test = np.array(pancakes_test,dtype='float32')
```

```python
train = np.concatenate((waffles_train,pancakes_train),axis=0)
```

```python
train.shape
```

```python
y = [0]*waffles_train.shape[0] + [1]*pancakes_train.shape[0]
```

```python
type(y)
```

```python
y = np.array(y,dtype='float32')
```

```python
y
```

```python
base_model = Xception(include_top = False, pooling='avg', input_shape = (224,224,3))
```

```python
base_model.summary()
```

```python
train_features = base_model.predict(train)
```

```python
train_features.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(solver='lbfgs',max_iter=1000)
```

```python
clf.fit(train_features,y)
```

```python
test = np.concatenate((waffles_test,pancakes_test),axis=0)
```

```python
y_test = [0]*waffles_test.shape[0] + [1]*pancakes_test.shape[0]
```

```python
y_test = np.array(y_test)
```

```python
test_features = base_model.predict(test)
```

```python
test_features.shape
```

```python
preds = clf.predict(test_features)
```

```python
(preds==y_test).sum()/y_test.shape[0]
```
