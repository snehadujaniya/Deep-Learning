```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from keras.applications.vgg16 import VGG16,decode_predictions,preprocess_input
from keras.preprocessing.image import load_img,img_to_array
%matplotlib inline
```

```python
os.chdir(r"E:\003 IBM CE\3. Artificial Intelligence\Mody\CNN\Pre-trained Convolution Networks")
```

```python
model=VGG16()
model.summary()
```

```python
model1=VGG16(include_top=False)
model1.summary()
```

```python
img=load_img("pizza.jpg",target_size=(224,224))
img_array=img_to_array(img)
img_array=np.expand_dims(img_array,axis=0)
img_arr_preprocess=preprocess_input(img_array)
```

```python
preds=model1.predict(img_arr_preprocess)
```
